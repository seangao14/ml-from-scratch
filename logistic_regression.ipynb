```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from src.logistic_regression import LogisticRegression
```

```python
X, y = make_blobs(100, centers=2)
plt.scatter(X[:,0], X[:,1])
```

```python
LR = LogisticRegression(lr=0.01, epochs=10000)
LR.fit(X,y)
```

```python
# graphing decision boundary: 

import numpy as np
x = np.linspace(-10,10,10000)
y = - (LR.bias + np.dot(LR.weights[0], x)) / LR.weights[1]
plt.plot(x, y)
plt.scatter(X[:,0], X[:,1])
```
